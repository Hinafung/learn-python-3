## if 条件判断

### if语句基本语法

if语句完整的语法结构：

```python
if <条件判断1>:
    <执行1>
elif <条件判断2>:
    <执行2>
elif <条件判断3>:
    <执行3>
else:
    <执行4>

```

if判断条件还可以简写，比如写：

```python
if x:
    print('True')
    
```
只要x是非零、非空、非false等，就判断为True，否则为False。

````python
x = 2
if x:
    print('True')
````

````python
x=2
if x:
    print('true')
````

````python
x = [1,2,3,4]
y = 2
if y in x:
    print(y+2)
````

````python
age = 22
if age >= 18:
    print("您的年龄是",age,"岁,您已成年")
````

````python
ticket = input("是否已买到票:")     # "是"代表有车票，"否"代表没有车票
knife = float(input("刀子的长度是:"))    # 刀子的长度，单位为cm

if ticket == "是":
    print("有车票，可以进站")
    if knife < 10:
        print("通过安检,请进站")
    else:
        print("没有通过安检,刀子的长度超过规定")
else:
    print("没有车票，不能进站")
````

## 循环语句

### while循环

Python中 while 语句用于循环执行程序，即在某条件下，循环执行某段程序，以处理需要重复处理的相同任务。其基本形式为
```python
while 判断条件：
    执行语句……```
    
执行语句可以是单个语句或语句块。判断条件可以是任何表达式，任何非零、或非空（null）的值均为true。

当判断条件假false时，循环结束。如果条件判断语句永远为 true，循环将会无限的执行下去

````python
i = 0  #变量一定要先赋值后使用
while i <= 10:
    print("abc")  #执行语句为单个语句
    i = i+1
````

### while循环嵌套

```python
while 条件1:
    满足条件1执行代码块1
    while条件2:
        满足条件2执行代码块2
    ......
```

````python
#  第一层循环
i = 1           
while i <= 5:
    print("*",end=' ')  #设置结束符为空,默认结束符是换行
    i += 1
````

````python
#  第一层循环
i = 1           
while i <= 5:
    print("*")  
    i += 1
````

````python
#在上一层循环的基础上再加一层循环
i = 1
while i <= 5:
    j = 1
    while j <= 5:
        print("*",end=' ')
        j += 1 
    print()
#     print("----第{}次大循环结束----".format(i))
    i += 1
````

````python
i = 1
while i<=5:
    j = 1
    while j <= i:   #和正方形的区别,不容易理解的时候可以在print函数例的end参数设置一个反斜杠帮助理解,正方形每一行都循环5次,三角形不是
        print("* ",end='')  
        j+=1
    print()
#     print("----第{}次大循环结束----".format(i))
    i+=1
````

````python
i = 1
while i<=5:
    j = 1
    while j <= i:   #和正方形的区别,不容易理解的时候可以在print函数例的end参数设置一个反斜杠帮助理解,正方形每一行都循环5次,三角形不是
        print("* ",end='/')  
        j+=1
    print()
    i+=1
````

#### break和continue
- break 结束整个循环
- continue 终止本次循环,继续下一次循环

注意:

- break/continue只能用在循环中，除此以外不能单独使用
- break/continue在嵌套循环中，只对最近的一层循环起作用

````python
words = '卸载王者荣耀'
for x in words:
    if x =="者":
        break
    print(x)
        
````

````python
words = '卸载王者荣耀'
for x in words:
    if x =="者":
        continue
    print(x)
````

````python
#continue
words = '卸载王者荣耀'
for x in words:
    print("-.-")
    if x =="者":
        continue
    print(x)
````

# pass语句

pass 是空语句，是为了保持程序结构的完整性。

pass 不做任何事情，一般用做占位语句。

````python
for letter in 'Python':
    if letter == 'h':
        print ('这是 pass 块')
    print ('当前字母 :', letter)
    
print ("Good bye!")
````

````python
for letter in 'Python':
    if letter == 'h':
        pass
        print ('这是 pass 块')
    else :
        print ('当前字母 :', letter)
    
print ("Good bye!")
````
